# SVM - SUPPORT VECTOR MACHINE

## IRIS DATA

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
X=iris.data[:,[2,3]]
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.svm import SVC
sc=StandardScaler()
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=1,test_size=0.3)
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

#### SVM LINEAL

```python
svm=SVC(kernel='linear',C=1,random_state=1)
svm.fit(X_train,y_train)
predicts=svm.predict(X_test)
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt

def plot_decision_region(X,y,classifier,test_idx=None,resolution=0.02):
    markers=('s','x','o','^','v')
    colors=('red','blue','lightgreen','gray','cyan')
    cmap=ListedColormap(colors[:len(np.unique(y))])

    x1_min, x1_max =X[:,0].min()- 1,X[:,0].max()+1
    x2_min, x2_max =X[:,1].min()- 1,X[:,1].max()+1
    xx1,xx2 =np.meshgrid(np.arange(x1_min,x1_max,resolution),
                         np.arange(x2_min,x2_max,resolution))
    Z=classifier.predict(np.array([xx1.ravel(),xx2.ravel()]).T)
    Z=Z.reshape(xx1.shape)
    plt.contourf(xx1,xx2,Z,alpha=0.3,cmap=cmap)
    plt.xlim(xx1.min(),xx1.max())
    plt.ylim(xx2.min(),xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y== cl,0],y=X[y==cl,1],
                    alpha=0.8,c=colors[idx],
                    marker=markers[idx], label=cl,
                    edgecolors='black')
    
    if test_idx:
        X_test,y_test =X[test_idx,:],y[test_idx]
        plt.scatter(X_test[:,0],X_test[:,1],c='white',edgecolors='black',alpha=0.3,
                    linewidths=1,marker='o',s=100,label='test_set')
```

```python
x_combined=np.vstack((X_train,X_test))
y_combined=np.hstack((y_train,y_test))
plot_decision_region(X_train,y_train,classifier=svm,test_idx=range(55,105))
plt.xlabel('petal lenght')
plt.ylabel('petal width')
plt.legend(loc='upper left')
plt.show()
```

```python
svm.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predicts)
```
